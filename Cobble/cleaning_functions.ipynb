```python
class Clean():
    list_of_dicts = []
    # Use on location, nested_categories
    def string_to_list(self, df, col):
        for i in range(len(df)):
            list_dict = json.loads(df[f'{col}'][i])
            list_of_dicts.append(list_dict)


    dicts = []
    # Use on location
    def list_to_dict(self, list_of_dicts):
        for i in range(len(list_of_dicts)):
            if len(list_of_dicts[i]) == 1:
                one_dict = list_of_dicts[i][0]
                dicts.append(one_dict)
            elif len(list_of_dicts[i]) > 1:
                two_dict = list_of_dicts[i][0]
                dicts.append(two_dict)
            elif len(list_of_dicts[i]) == 0:
                dicts.append('NA')
             
    lists_of_cats = []

    # Use on nested_categories after running string_to_list.
    def dict_to_list(self, list_of_dicts):
        for i in range(len(list_of_dicts)):
            list_of_cat = []

            for j in range(len(list_of_dicts[i])):
                cat = list(list_of_dicts[i][j].values())[0]
                list_of_cat.append(cat)

            lists_of_cats.append(list_of_cat)
```
