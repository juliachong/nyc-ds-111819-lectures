```python
def get_url(url):
    response = requests.get(url)
    # function passes through url1, url2 etc and puts it thru parser
    soup = BS(response.text, 'html.parser')
    urls = soup.find_all('div', class_='lister-list')
    # soup.find_all returns the overarching lister-list class (list of ALL movies on page)
    return urls
```

```python
# GETS URL AND PARSES THROUGH TO WHERE WE WANNA BE
def get_movie_id(url_list):
    ids = []
    for result in url_list:
        for x in range(0, 250):
            try:
                i = (result.find_all('img')[x])
                ids.append(i['data-tconst'])
            except:
                continue
        return ids
```
