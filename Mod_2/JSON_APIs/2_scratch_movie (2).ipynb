```python
# GETS URL AND PARSES THROUGH TO WHERE WE WANNA BE

def get_movie_id(url_list):
    ids = []
    for result in url_list:
        for x in range(0, 250):
            try:
                i = (result.find_all('img')[x])
                ids.append(i['data-tconst'])
            except:
                continue
        return ids
```
