```python
import csv
our_data = []
with open('data.csv') as f:
    # with data.csv which we are opening stored as 'f'
    reader = csv.DictReader(f)
    # csv is the package you are importing, .DictReader is part of the package
    for item in reader:
        our_data.append(item)
```

```python
# * Find by name - Takes in a string that represents the name of an album. 
# * Should return a dictionary with the correct album, or return `None`.

def find_my_name(name):
    for album in our_data:
        if album['album'] == name:
            return album
        else:
            return 'None'

        
find_my_name("In An Aeroplane Over The Sea")
```

```python
def find_by_rank(rank):
    for rank_variable in our_data:
        # rank_variable represents iterating through each data set in our_data
        if rank_variable['number'] == rank:
        # we call the current data set in which we are iterating through
        # and pass through the key 'number' to obtain the rank of the data set
        # we then pass it through a conditional : if this rank of the data set
        # is equal to the rank that we passed through,
            return '# ' + rank, rank_variable['album']

find_by_rank(str(5))
        
```

```python
def find_year(year):
    years = []
    for dic in our_data:
        if str(year) == dic['year']:
            years.append(dic)
        return years
find_year(1965)
```
