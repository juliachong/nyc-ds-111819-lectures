```python
# HOW TO CREATE A GRID OF SUBPLOTS WITH ONE LINE OF CODE!!!
x = np.linspace(-10, 10, 101)
# generates a set of 101 data points between -10 and 10 

fig, axes = plt.subplots(ncols=2, nrows=3, figsize = (10,10))
# creates subplots containing 2 columns and 3 rows of graphs

for n in range(1, 9):
    # for each number in the set 1 through 9 (1, 2, 3, 4, 5, 6, 7, 8, 9)
    row = (n-1)//2
    # row index is (n-1) / 2:
    # (0, .5, 1, 1.5, 2, 2.5, 3, 3.5 ...)
    col = n%2-1
    # col index is (n/2)-1
    # (-0.5, 0, .5, 1, 1.5, 2, 2.5, 3, 3.5 ...)
    ax = axes[row][col]
    # ax gives the row and column indices of each axes
    # [0, -.5], [.5, 0], [1, .5]
    y = [xi**n for xi in x]
    # the y equation is x^nth power for each value in the x set.
    ax.plot(x,y)
    # plots x against y for each graph
    ax.set_title('x^{}'.format(n))
    
```
